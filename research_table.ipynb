```python
import matplotlib.pyplot as plt

# Data from the provided table
tasks = [
    "Project planning & specs",
    "Data collection & consent",
    "Data cleaning & preprocessing",
    "Feature engineering",
    "Model selection & design",
    "Model training & tuning",
    "Evaluation & validation",
    "UI prototype & integration",
    "User testing & feedback",
    "Deployment & monitoring",
    "Documentation & report",
    "Buffer / project management"
]
start_weeks = [1, 2, 4, 6, 7, 8, 10, 11, 13, 14, 12, 1]
durations = [2, 3, 2, 2, 2, 3, 2, 2, 2, 2, 4, 16]

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 8))

# Plot horizontal bars for each task with the same color
for i, task in enumerate(tasks):
    ax.barh(task, durations[i], left=start_weeks[i], color="#2196F3")  # Uniform blue color

# Customize the chart
ax.set_xlabel("Weeks")
ax.set_title("MSBTE Result Automation Gantt Chart")
ax.set_xlim(0, 17)  # Set x-axis limit to cover 16 weeks + buffer
ax.grid(True, axis="x", linestyle="--", alpha=0.7)

# Invert y-axis to have tasks listed top-down
ax.invert_yaxis()

# Display the chart
plt.tight_layout()
plt.show()
```
